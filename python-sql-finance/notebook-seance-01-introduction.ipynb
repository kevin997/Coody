## 3. Opérateurs arithmétiques

### 3.1 Opérations de base

```python
# Calculs financiers simples
prix_achat = 100
prix_vente = 120

profit = prix_vente - prix_achat
rendement = (profit / prix_achat) * 100

print(f"Prix d'achat : {prix_achat} FCFA")
print(f"Prix de vente : {prix_vente} FCFA")
print(f"Profit : {profit} FCFA")
print(f"Rendement : {rendement}%")
```

### 3.2 Opérations avancées

```python
# Puissance (pour intérêts composés)
capital = 10000
taux = 1.05  # 5% de croissance
annees = 3

valeur_future = capital * (taux ** annees)

print(f"Capital initial : {capital:,.0f} FCFA")
print(f"Après {annees} ans : {valeur_future:,.2f} FCFA")
```

## 4. Input/Output interactif

### 4.1 Demander des informations à l'utilisateur

```python
# Calculateur de profit interactif
nom = input("Quel est votre nom ? ")
print(f"Bonjour {nom}!")

prix_achat = float(input("Prix d'achat (FCFA) : "))
prix_vente = float(input("Prix de vente (FCFA) : "))

profit = prix_vente - prix_achat
rendement = (profit / prix_achat) * 100

print(f"\nRésultat pour {nom} :")
print(f"Profit : {profit:,.2f} FCFA")
print(f"Rendement : {rendement:.2f}%")
```

## 5. Application : Calculateur d'intérêts simples

**Formule** : Intérêts = Principal × Taux × Temps

```python
# Calculateur d'intérêts simples
print("=" * 50)
print("CALCULATEUR D'INTÉRÊTS SIMPLES")
print("=" * 50)

# Entrées
capital = float(input("\nCapital initial (FCFA) : "))
taux_pourcent = float(input("Taux d'intérêt annuel (%) : "))
duree = int(input("Durée (années) : "))

# Calculs
taux = taux_pourcent / 100
interets = capital * taux * duree
montant_final = capital + interets

# Résultats
print("\n" + "=" * 50)
print("RÉSULTATS")
print("=" * 50)
print(f"Capital initial       : {capital:>15,.0f} FCFA")
print(f"Taux annuel           : {taux_pourcent:>15.2f}%")
print(f"Durée                 : {duree:>15} ans")
print("-" * 50)
print(f"Intérêts gagnés       : {interets:>15,.0f} FCFA")
print(f"Montant final         : {montant_final:>15,.0f} FCFA")
print("=" * 50)
```

**Solution Exercice 3** (à dévoiler après tentative)

```python
# Solution
print("CALCULATEUR DE VALEUR DE PORTEFEUILLE\n")

# Action 1
nom1 = input("Action 1 - Nom : ")
prix1 = float(input("Action 1 - Prix : "))
qte1 = int(input("Action 1 - Quantité : "))

# Action 2
nom2 = input("\nAction 2 - Nom : ")
prix2 = float(input("Action 2 - Prix : "))
qte2 = int(input("Action 2 - Quantité : "))

# Action 3
nom3 = input("\nAction 3 - Nom : ")
prix3 = float(input("Action 3 - Prix : "))
qte3 = int(input("Action 3 - Quantité : "))

# Calculs
valeur1 = prix1 * qte1
valeur2 = prix2 * qte2
valeur3 = prix3 * qte3
valeur_totale = valeur1 + valeur2 + valeur3

# Affichage
print("\n" + "="*50)
print("COMPOSITION DU PORTEFEUILLE")
print("="*50)
print(f"{nom1:<15} : {valeur1:>15,.2f} FCFA")
print(f"{nom2:<15} : {valeur2:>15,.2f} FCFA")
print(f"{nom3:<15} : {valeur3:>15,.2f} FCFA")
print("-"*50)
print(f"{'TOTAL':<15} : {valeur_totale:>15,.2f} FCFA")
print("="*50)
```
